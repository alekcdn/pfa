## Задача 2

На складе лежат разные фрукты в разном количестве.
Нужно написать функцию, которая на вход принимает 
любое количество названий фруктов и их количество, 
а возвращает общее количество фруктов на складе.

```python
fruits = '''
Яблоки - 10
Груши - 25
Бананы - 55
'''
import re

def total_(a):
    return sum([int(i.group()) for i in re.finditer(r'\d+', a)])

print('Всего фруктов: ', total_(fruits))
```

## Задача 3

Дан список с затратами на рекламу. Но в данных есть ошибки, некоторые затраты имеют отрицательную величину. 
Удалите такие значения из списка и посчитайте суммарные затраты
[100, 125, -90, 345, 655, -1, 0, 200]
Используйте list comprehensions

```python
array = [100, 125, -90, 345, 655, -1, 0, 200]
new_array = [a for a in array if a >= 0]
print(sum(new_array))
```

### 4.2 Найдите выручку компании в зависимости от месяца
Для этого напишите функцию, которая на вход принимает список с датами и список с выручкой, а на выходе словарь, где ключи - это месяцы, а значения - это выручка.
Используйте аннотирование типов.

```python
def sum_(date: list, money: list) -> dict:
    res = {}
    for d, m in zip(date, money):
        if d[5:7] in res.keys():
            res[d[5:7]] = res[d[5:7]] + m
        else:
            res[d[5:7]] = m
    return res
 
print(sum_(date, money))
```
